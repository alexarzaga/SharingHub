```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
media_df = pd.read_csv('files/country_media.csv')
media_df = media_df.dropna(thresh=5)
```

```python
media_df
```

```python
comp = plt.scatter(x=media_df['Internet'], 
            y=media_df['Computer'],
           color='red',
           marker="o",
           label='Computer')
cell = plt.scatter(x=media_df['Internet'],
           y=media_df['Cellphone'],
           color='blue',
           marker='s',
           label='Cellphone')
tv = plt.scatter(x=media_df['Internet'],
           y=media_df['TV'],
           color='green',
           marker='^',
           label='TV')
plt.xlabel('Internet')
plt.ylabel('Media Device')
plt.title('Media Device Correlation with Internet')
plt.legend((comp, cell, tv), ('Computer', 'Cellphone', 'TV'))
plt.show()
```

# What Do You See?
## Media Device Ownership in Correlation with Internet Ownership
The data we are looking at is the proportion of housholds that own the three various types of media devices (Computer, Cellphone, and TV), in comparison to the proportion of housholds that have internet. There is a strong positive correlation between households that own computer and have internet. While, celphones and TVs have a slightly different correlation. There may be some mathematical manipulation that can show us the true correlation of the cellphone and tvs with the internet, but this more importantly shows us the different correlations across all these devices have the internet ownership without manipulation of the data.
